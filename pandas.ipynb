```python
import pandas as pd
class ReadData:
    def __init__(self, data = None, delimiter=None):
        self.data = pd.read_csv(data, delimiter=delimiter)
    def display(self):
        print(self.data.head())
read_data = ReadData("C:\\Users\\chakr\\OneDrive\\Desktop\\data.txt.txt", delimiter="\t")
read_data.display()
groupby_df = read_data.data.groupby('Company')['Room number'].sum()
print(groupby_df)
```
